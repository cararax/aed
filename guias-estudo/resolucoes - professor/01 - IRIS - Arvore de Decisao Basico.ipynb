Importar bibliotecas necessárias:
sklearn.datasets para carregar datase IRIS
train_test_split para dividir dataset em subconjuntos de treino e teste
DecisionTreeClassifier para instanciar modelo de árvore de decisão
accuracy_score para avaliação do modelo


```python
# ============================================================
# 1. IMPORTANDO AS BIBLIOTECAS NECESSÁRIAS
# ============================================================
# sklearn.datasets: contém datasets prontos para aprendizado (como o Iris)
from sklearn.datasets import load_iris
# train_test_split: divide os dados em treino (aprender) e teste (avaliar)
from sklearn.model_selection import train_test_split
# DecisionTreeClassifier: algoritmo de árvore de decisão para classificação
from sklearn.tree import DecisionTreeClassifier
# accuracy_score: métrica que mede o % de acertos do modelo
from sklearn.metrics import accuracy_score
```

Carregar Dados:
iris = load_iris()


```python
# ============================================================
# 2. CARREGANDO O DATASET IRIS
# ============================================================
# O dataset Iris contém 150 amostras de flores com 4 características:
# - comprimento da sépala, largura da sépala, comprimento da pétala, largura da pétala
# O objetivo é classificar a espécie (Setosa, Versicolor, Virginica)
iris = load_iris()
```

Preparar X(características) e y (classe):
X = iris.data, y = iris.target


```python
# ============================================================
# 3. SEPARANDO FEATURES (X) E TARGET (y)
# ============================================================
# X = matriz de características (features) — são as "entradas" do modelo
#     iris.data contém as 4 medidas das pétalas/sépalas para cada flor
X = iris.data

# y = vetor alvo (target) — é a "resposta" que queremos prever
#     iris.target contém o código da espécie (0=Setosa, 1=Versicolor, 2=Virginica)
y = iris.target

# 💡 DICA: Em problemas reais, você carregaria seus dados de um CSV com pd.read_csv()
#    e depois faria: X = df[['coluna1', 'coluna2']] e y = df['target']
```

Dividir Dados entre treino e teste:
X_train recebe características dos indíviduos para treino (70%);
X_test recebe características dos indíviduos para teste (30%);
y_train recebe as classes dos indíviduos para treino (70%);
y_test recebe as classes dos indíviduos para teste (30%);


```python
# ============================================================
# 4. DIVIDINDO OS DADOS EM TREINO E TESTE
# ============================================================
# ⚠️ NUNCA teste um modelo com os mesmos dados que ele usou para aprender!
#    Senão ele apenas "decora" as respostas (overfitting).
#
# Aqui separamos:
#   - 70% dos dados para TREINO (o modelo aprende com esses)
#   - 30% para TESTE (o modelo tenta acertar sem ter visto antes)
#
# random_state=42: fixa uma "semente" para a divisão ser sempre igual
#                  (resultados reproduzíveis = mesma resposta toda vez)
# stratify=y:      mantém a proporção de cada espécie no treino e no teste
#                  (importante quando as classes são desbalanceadas!)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

Instanciar Modelo. Para árvore de decisão é o seguinte.

```python
# ============================================================
# 5. INSTANCIANDO O MODELO (ÁRVORE DE DECISÃO)
# ============================================================
# DecisionTreeClassifier é um algoritmo que cria uma "árvore de perguntas"
# para classificar os dados. Cada nó pergunta algo sobre uma feature,
# e cada ramo leva a uma resposta diferente.
#
# random_state=42: garante que a árvore sempre monte a mesma estrutura
model = DecisionTreeClassifier(random_state=42)

# 💡 DICA: Outros hiperparâmetros importantes (veremos adiante):
#   - max_depth:  profundidade máxima da árvore (evita overfitting)
#   - criterion:  'gini' (padrão) ou 'entropy' (medida de impureza)
```

Treinar Modelo (usa apenas 70% dos dados).

```python
# ============================================================
# 6. TREINANDO O MODELO (FIT)
# ============================================================
# O método .fit() é ONDE O APRENDIZADO ACONTECE.
# O modelo analisa os dados de treino (X_train) e suas respostas (y_train)
# para descobrir padrões — como "se pétala > 2.5cm então é Versicolor".
#
# 🧠 Pense no fit() como o "período de estudos" do modelo.
model.fit(X_train, y_train)

# Esta mensagem é só um aviso visual — o treino já aconteceu na linha acima
print("Modelo treinado com sucesso!")
```

Fazer Previsões. Analise apenas as características dos 30% restantes do dataset.Sem verificar suas classes.


```python
# ============================================================
# 7. FAZENDO PREVISÕES (PREDICT)
# ============================================================
# Agora o modelo vai tentar classificar as flores que ele NUNCA VIU antes (X_test).
# Ele usa apenas as características (medidas) — sem olhar a resposta verdadeira!
#
# 🧠 Pense no predict() como a "prova" do modelo — ele precisa acertar sem colar.
predictions = model.predict(X_test)

# predict_proba() também existe e retorna a probabilidade de cada classe.
# Exemplo: [0.95, 0.05, 0.0] = 95% de chance de ser Setosa
```

Avaliação do modelo treinado usando a métrica acurácia:


```python
# ============================================================
# 8. AVALIANDO O MODELO (ACURÁCIA)
# ============================================================
# Acurácia = (previsões corretas) / (total de previsões) * 100
# É a métrica mais simples — mede o percentual de acertos.
#
# ⚠️ CUIDADO: Acurácia pode ser enganosa se as classes forem desbalanceadas!
#    Ex: se 95% das flores são Setosa, um modelo que chuta "Setosa" sempre
#    teria 95% de acurácia sem aprender nada!
accuracy = accuracy_score(y_test, predictions)

# Exibe o resultado (0.0 a 1.0) — multiplique por 100 para percentual
print(accuracy)

# 💡 DICA: Para uma avaliação mais completa, use também:
#   - confusion_matrix(y_test, predictions) → mostra erros por classe
#   - classification_report(y_test, predictions) → precision, recall, f1
```
