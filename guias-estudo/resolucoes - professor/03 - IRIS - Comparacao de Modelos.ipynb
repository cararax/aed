```python
# ============================================================
# 1. IMPORTANDO AS BIBLIOTECAS
# ============================================================
from sklearn.datasets import load_iris         # Dataset Iris (flores)
from sklearn.model_selection import train_test_split  # Divisão treino/teste
from sklearn.tree import DecisionTreeClassifier # Árvore de Decisão
from sklearn.metrics import accuracy_score     # Métrica de avaliação
import matplotlib.pyplot as plt                # Plotagem de gráficos
from sklearn.tree import plot_tree             # Visualização da árvore

# 💡 DICA: Todas essas bibliotecas são do ecossistema scikit-learn + matplotlib.
#    Se estiver no Google Colab, já vêm instaladas!
```

```python
# ============================================================
# 2. CARREGANDO E EXPLORANDO O DATASET IRIS
# ============================================================

iris = load_iris()

# X = FEATURES (características): medidas das pétalas e sépalas
X = iris.data

# y = TARGET (rótulos): espécies (0=Setosa, 1=Versicolor, 2=Virginica)
y = iris.target

# Vamos explorar os dados antes de modelar:
print("Nomes das Features:", iris.feature_names)
# ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

print("Nomes das Classes (Target):", iris.target_names)
# ['setosa' 'versicolor' 'virginica']

# Verificando as primeiras amostras:
print("\nPrimeiras 5 linhas das Features (X):\n", X[:5])
print("\nPrimeiras 5 linhas do Target (y):\n", y[:5])

# 🧠 Observação: Note que cada linha de X corresponde a uma flor,
#    e o valor em y na mesma posição é a espécie daquela flor.
```

```python
# ============================================================
# 3. DIVIDINDO EM TREINO (70%) E TESTE (30%)
# ============================================================
# ⚠️ SEPARAÇÃO É CRÍTICA! Usar os mesmos dados para treino e teste
#    faz o modelo apenas "decorar" em vez de "aprender".
#
# test_size=0.3 → 30% das flores vão para teste, 70% para treino
# random_state=42 → semente fixa para resultados idênticos toda vez
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nTamanho do conjunto de treino (X_train): {X_train.shape[0]} amostras")
print(f"Tamanho do conjunto de teste (X_test): {X_test.shape[0]} amostras")
# Esperado: 105 treino, 45 teste (150 total)
```

```python
# ============================================================
# 4. INSTANCIANDO A ÁRVORE DE DECISÃO
# ============================================================
# Ainda NÃO treinamos — apenas criamos o objeto do modelo.
# random_state=42 garante que a mesma árvore seja gerada toda vez.
model = DecisionTreeClassifier(random_state=42)
```

```python
# ============================================================
# 5. TREINAMENTO (FIT) — O APRENDIZADO ACONTECE AQUI!
# ============================================================
# O .fit() recebe:
#   - X_train: as medidas (features) das flores de treino
#   - y_train: as espécies (rótulos) dessas flores
#
# Internamente, o algoritmo encontra os melhores pontos de corte
# nas features para separar as espécies com a maior pureza possível.
model.fit(X_train, y_train)

print("\nModelo treinado com sucesso!")
# Se não houve erro, o modelo está pronto para fazer previsões!
```

```python
# ============================================================
# 6. PREVISÕES (PREDICT) — A "PROVA" DO MODELO
# ============================================================
# O modelo classifica as flores de TESTE (que nunca viu).
# Usa APENAS as características (X_test) — sem olhar as respostas!
predictions = model.predict(X_test)

# Vamos conferir os resultados das primeiras 10 flores:
print("\nAlgumas previsões do modelo no conjunto de teste:")
for i in range(10):
    predicted_species = iris.target_names[predictions[i]]
    actual_species = iris.target_names[y_test[i]]
    marcador = "✅" if predictions[i] == y_test[i] else "❌"
    print(f"Flor {i+1}: Previsão: {predicted_species} | Real: {actual_species} {marcador}")
```

```python
# ============================================================
# 7. AVALIAÇÃO E DETECÇÃO DE OVERFITTING
# ============================================================

# --- Acurácia no TESTE (dados não vistos durante o treino) ---
test_accuracy = accuracy_score(y_test, predictions)

# --- Acurácia no TREINO (para comparar) ---
train_predictions = model.predict(X_train)
train_accuracy = accuracy_score(y_train, train_predictions)

print(f"\n--- Avaliação do Modelo ---")
print(f"Acurácia de TREINO: {train_accuracy:.2f}")
print(f"Acurácia de TESTE:  {test_accuracy:.2f}")

# 🧠 OVERFITTING:
# Se a acurácia de treino é muito maior que a de teste, o modelo "decorou"
# em vez de aprender. É como um aluno que sabe a prova antiga de cor
# (treino) mas não consegue resolver questões novas (teste).
if train_accuracy > test_accuracy + 0.01:
    print("\n⚠️  Alerta de Overfitting!")
    print("   Acurácia de treino > Acurácia de teste.")
    print("   O modelo pode estar decorando os dados de treino.")
else:
    print("\n✅ O modelo parece estar generalizando bem.")

print("\n" + "="*60)
print("💡 COMO RESOLVER OVERFITTING EM ÁRVORES DE DECISÃO:")
print("   - max_depth: limitar a profundidade da árvore")
print("   - min_samples_leaf: mínimo de amostras por folha")
print("   - min_samples_split: mínimo para dividir um nó")
print("="*60)

# --- Vamos testar com PODA (max_depth=3) ---
model_pruned = DecisionTreeClassifier(max_depth=3, random_state=42)
model_pruned.fit(X_train, y_train)

train_predictions_pruned = model_pruned.predict(X_train)
test_predictions_pruned = model_pruned.predict(X_test)

train_accuracy_pruned = accuracy_score(y_train, train_predictions_pruned)
test_accuracy_pruned = accuracy_score(y_test, test_predictions_pruned)

print(f"\n--- Modelo com max_depth={model_pruned.max_depth} (PODADO) ---")
print(f"Acurácia de TREINO (podado): {train_accuracy_pruned:.2f}")
print(f"Acurácia de TESTE (podado):  {test_accuracy_pruned:.2f}")

if train_accuracy_pruned > test_accuracy_pruned + 0.01:
    print("A diferença entre treino e teste ainda existe, mas reduziu.")
else:
    print("✅ Com a poda, a diferença entre treino e teste DIMINUIU!")
    print("   O modelo agora GENERALIZA melhor para dados novos.")

# 💡 DICA: Teste diferentes valores de max_depth (1, 2, 3, 5, None)
#    e veja como o equilíbrio viés-variância se comporta!
```

```python
# ============================================================
# 8. COMPARANDO DIFERENTES ALGORITMOS DE CLASSIFICAÇÃO
# ============================================================
# Não existe um único algoritmo "melhor" para tudo.
# Cada dataset e problema pede uma abordagem diferente.
# Vamos comparar 6 algoritmos diferentes no MESMO dataset!

print("--- Comparando Diferentes Modelos de Classificação ---")

# Importando outros algoritmos do sklearn
from sklearn.neighbors import KNeighborsClassifier      # K-Vizinhos
from sklearn.linear_model import LogisticRegression      # Regressão Logística
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis  # LDA
from sklearn.naive_bayes import GaussianNB              # Naive Bayes Gaussiano
from sklearn.svm import SVC                             # Máquina de Vetores de Suporte
from sklearn.tree import DecisionTreeClassifier          # Árvore de Decisão (já vimos)

# --- Dicionário com os modelos ---
# Cada modelo tem sua própria "forma de aprender":
models = {
    "🌳 Árvore de Decisão (max_depth=3)": DecisionTreeClassifier(max_depth=3, random_state=42),
    "📏 K-Nearest Neighbors (k=5)": KNeighborsClassifier(n_neighbors=5),
    "📈 Regressão Logística": LogisticRegression(random_state=42, max_iter=200),
    "📊 LDA (Análise Discriminante)": LinearDiscriminantAnalysis(),
    "🧮 Naive Bayes (Gaussiano)": GaussianNB(),
    "🎯 SVM (Support Vector Machine)": SVC(random_state=42)
}

# --- Treinar e avaliar cada modelo ---
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)          # Passo 1: treinar
    predictions = model.predict(X_test)  # Passo 2: prever
    accuracy = accuracy_score(y_test, predictions)  # Passo 3: avaliar
    results[name] = accuracy
    print(f"  {name}: {accuracy:.2f}")

# --- Resultados finais ---
print("\n" + "="*60)
print("📊 RESUMO — ACURÁCIA POR ALGORITMO")
print("="*60)
# Ordena do melhor para o pior
for name, accuracy in sorted(results.items(), key=lambda x: x[1], reverse=True):
    estrelas = "⭐" * int(accuracy * 5)  # Visual: mais estrelas = melhor
    print(f"  {name}: {accuracy:.2f} {estrelas}")

print("\n🧠 CONCLUSÕES DA COMPARAÇÃO:")
print("  • Todos os algoritmos tiveram desempenho similar no Iris,")
print("    porque é um dataset relativamente simples e bem separável.")
print("  • Em datasets mais complexos, as diferenças seriam MAIORES.")
print("  • A escolha do modelo depende de:")
print("    - Tamanho e natureza dos dados")
print("    - Interpretabilidade necessária")
print("    - Tempo de treino disponível")
print("    - Tipo de problema (linear, não-linear, etc.)")
print("  • ⚠️ NUNCA escolha um modelo só pela acurácia —")
print("    considere também tempo, interpretabilidade e robustez!")

# 💡 DICA: Experimente adicionar outros modelos como:
#   - RandomForestClassifier (ensemble de árvores)
#   - GradientBoostingClassifier
#   - MLPClassifier (redes neurais simples)
```
