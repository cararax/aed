Importar bibliotecas necessárias:
sklearn.datasets para carregar datase IRIS
train_test_split para dividir dataset em subconjuntos de treino e teste
DecisionTreeClassifier para instanciar modelo de árvore de decisão
accuracy_score para avaliação do modelo


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

Carregar Dados:
iris = load_iris()


```python
iris = load_iris()
```

Preparar X(características) e y (classe):
X = iris.data, y = iris.target


```python
X = iris.data
y = iris.target
```

Dividir Dados entre treino e teste:
X_train recebe características dos indíviduos para treino (70%);
X_test recebe características dos indíviduos para teste (30%);
y_train recebe as classes dos indíviduos para treino (70%);
y_test recebe as classes dos indíviduos para teste (30%);


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

Instanciar Modelo. Para árvore de decisão é o seguinte.

```python
model = DecisionTreeClassifier(random_state=42)
```

Treinar Modelo (usa apenas 70% dos dados).

```python
model.fit(X_train, y_train)

print("Modelo treinado com sucesso!")
```

Fazer Previsões. Analise apenas as características dos 30% restantes do dataset.Sem verificar suas classes.


```python
predictions = model.predict(X_test)
```

Avaliação do modelo treinado usando a métrica acurácia:


```python
accuracy = accuracy_score(y_test, predictions)

print(accuracy)
```
